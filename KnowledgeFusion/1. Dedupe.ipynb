## 2. 定距离函数

```python
#定义相似度函数
import difflib
# import jieba
import Levenshtein

def Similarity(str1,str2,distance_type):
    if distance_type == "difflib":
        seq = difflib.SequenceMatcher(None, str1,str2)
        ratio = seq.ratio()
        return ratio
    if distance_type == "edit":
        # 3. 编辑距离，描述由一个字串转化成另一个字串最少的操作次数，在其中的操作包括 插入、删除、替换
        sim = Levenshtein.distance(str1, str2)
        return sim
    if distance_type == "lev":
        # 4.计算莱文斯坦比
        sim = Levenshtein.ratio(str1, str2)
        return sim
    if distance_type == "jaro":
        # 5.计算jaro距离
        sim = Levenshtein.jaro(str1, str2 )
        return sim
    if distance_type == "jw":
        # 6. Jaro–Winkler距离
        sim = Levenshtein.jaro_winkler(str1 , str2 )
        return sim
```

```python
#相似度函数测试

str1 = "中文名"
str2 = "外文名"
  
# 1. difflib
print ('difflib similarity: ', Similarity(str1,str2,"difflib"))
# 2. edit
print ('edit similarity: ', Similarity(str1,str2,"edit"))
# 3. lev
print ('lev similarity: ', Similarity(str1,str2,"lev"))
# 4. jaro
print ('jaro similarity: ', Similarity(str1,str2,"jaro"))
# 5. jw
print ('jw similarity: ', Similarity(str1,str2,"jw"))
```

```python
#思路：
#遍历“属性表头”的列表，找到距离相近的“属性表头”组
#复制原表表头到一个列表中，找到“同组的属性表头”，将该“属性表头”的名称保存到一个列表中，然后删除该“属性表头”，
search_list = np.array(data_new.columns)
print(search_list[3:])
print(len(search_list))
```

### 3.1 初步贪婪聚类

```python
#用diff距离，对属性进行贪婪聚类
search_list = np.array(data_new.columns)
diffSimilar_list = []
for i,col_1 in enumerate(search_list[4:]):
    flag = 0
    for te in diffSimilar_list:
        if col_1 in te:
            flag = 1
            break
    if flag == 1:
        continue
    similar = []
    similar.append(col_1)
    for j,col_2 in enumerate(search_list[4+i+1:]):
        if (len(col_2)> 2 and len(col_1)> 2 and Similarity(col_1,col_2,"edit") < 2) or ( Similarity(col_1,col_2,"difflib") > 0.6):
            similar.append(col_2)
            
    diffSimilar_list.append(similar)
    print(similar)
for i,j in enumerate(diffSimilar_list):
     print(i,j)
```

## 5. 定义“非空率降序属性类型表”生成函数
