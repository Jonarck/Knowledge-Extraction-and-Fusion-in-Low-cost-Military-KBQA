## 2.3 正式融合

```python
#以武器名称为索引并去重
struData.set_index(["weapon"], inplace=True)
print("去重前索引长度（武器总数）：",len(struData.index))
struData = struData.reset_index().drop_duplicates(subset='weapon', keep='first').set_index('weapon')
print("去重后索引长度（武器总数）：",len(struData.index))
```

```python
unstruData.set_index(["weapon"], inplace=True)
unstruData =  unstruData.reset_index().drop_duplicates(subset='weapon', keep='first').set_index('weapon')
```

```python
print(np.array(struData.index))
print(np.array(unstruData.index))
```

```python
#关键是分情况讨论
#1.定义被融合的属性列名集
merge_prop_name = ['别名', '时间', '国家', '特点', '作用']
weapon_s = np.array(struData.index)
weapon_u = np.array(unstruData.index)

#2.拷贝接受修改的结构化数据
struDataC = pd.DataFrame.copy(struData,deep = True)
print(struDataC)
```

```python
increase = [0,#别名--0
            0,#时间--1
            0,#国家--2
            0,#特点--3
            0]#作用--4
add = [0,#别名--0
       0,#时间--1
       0,#国家--2
       0,#特点--3
       0]#作用--4
same = 0
#把用非结构化提炼的数据插入结构化数据表
for index_u, row_u in unstruData.iterrows():
    #已知全部非结构化数据中的武器都被结构化数据中包含
    print("开始融合：",index_u," 的武器属性")
    for p_n,prop in enumerate(merge_prop_name):
        #0.情况零：若该武器的该属性为空，则无需添加到结构化数据表中，直接进入该武器下一个属性的判断
        if row_u[prop] == "":
            continue
        #1.情况一：专门对国家而言，若结构化数据中已有国家，则不需将非结构化数据中的国家填入，否则将非结构化数据中的国家填入
        if prop == "国家":
            if struData[prop][index_u] != "":
                same += 1
                continue
            struDataC[prop][index_u] += row_u[prop]
            increase[p_n] += 1
            continue
        #2.情况二：对非国家属性而言，若结构化数据中有内容，则判断内容是否相同将此处内容加上（“其他”+“属性名”：“属性值”；）
        if struData[prop][index_u] != "":
            #改进：用冒号和分号分开，struData[prop][index_u]字符串，依次判定字符距离，决定是否添加
            judge_list = re.split('[:;,：；，]', struData[prop][index_u])
            if row_u[prop] in judge_list:#若有内容重复，则不添加
                same += 1
                continue
            struDataC[prop][index_u] += ("其他"+prop+":"+row_u[prop]+";")
            add[p_n] += 1
            continue
        #2.情况三：对非国家属性而言，若结构化数据中无内容，则再分情况加上（“属性名”：“属性值”；）
        #时间：
        if prop == '时间':
            struDataC[prop][index_u] += ("重要"+prop+":"+row_u[prop]+";")
            increase[p_n] += 1
            continue
        #别名：
        if prop == '别名':
            struDataC[prop][index_u] += ("常见"+prop+":"+row_u[prop]+";")
            increase[p_n] += 1
            continue
        #特点与作用：
        struDataC[prop][index_u] += (prop+":"+row_u[prop]+";")
        increase[p_n] += 1
```

```python
print("内容扩充条数：",dict(zip(merge_prop_name,add)))
print("内容新增条数：",dict(zip(merge_prop_name,increase)))
print("避开的重复内容项目数：",same)

rate_list = []

for prop in merge_prop_name:
    num_i = 0
    for index_u, row_u in struData.iterrows():
        if row_u[prop] != "":
            num_i += 1
    num_all = len(struData[prop])
    rate = num_i/num_all
    rate_list.append(str(round(round(rate,2)*100,2))+"%")
print("相关列原始非空率",dict(zip(merge_prop_name,rate_list)))

rate_listC = []
for prop in merge_prop_name:
    num_i = 0
    for index_u, row_u in struDataC.iterrows():
        if row_u[prop] != "":
            num_i += 1
    num_all = len(struDataC[prop])
    rate = num_i/num_all
    rate_listC.append(str(round(round(rate,2)*100,2))+"%")
print("相关列原始非空率",dict(zip(merge_prop_name,rate_listC)))
```
